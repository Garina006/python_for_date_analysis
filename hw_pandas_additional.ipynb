6. Трансформуйте колонки 'Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків' у тип даних float. УВАГА! Перед трансформацією подивіться, які дані містять ці колонки. Аби успішно трансформувати тип даних, треба спочатку привести рядки до того виду, який буде підходити, аби потім рядок зробити типом float.

```python
df_new = df.copy()
```

```python
#Перетворюємо всі значення NaN на 0 в колонці 'Зарплата у $$$ за місяць, лише ставка після сплати податків'. 
#Колонка 'Ваш вік' не містить значення NaN, про що ми дізнались раніше, тому вона не потребує перетворення
df_new['Зарплата у $$$ за місяць, лише ставка після сплати податків'].fillna(0, inplace = True)
```

```python
#Замінюємо кому на крапку в усіх значеннях для подальшого перетворення в тип даних float
df_new['Ваш вік'] = df_new['Ваш вік'].astype(str).str.replace(',', '.', regex=True).astype(float)
df_new['Зарплата у $$$ за місяць, лише ставка після сплати податків'] = (df_new['Зарплата у $$$ за місяць, лише ставка після сплати податків']
                                                                         .astype(str)
                                                                         .str.replace(',', '.', regex=True)
                                                                         .astype(float))
```

```python
#Оскільки маємо в даних два різних формати для запису дат, формуємо різні серії, які заповнюються значеннями, що відповідають формату,
#вказанному у функції to_datetime. Значення, що не відповідають формату, заповнюються пустими значеннями
date_1 = pd.to_datetime(df_new['Timestamp'], dayfirst=True, errors='coerce', format='%m.%d.%Y %H:%M:%S')
date_2 = pd.to_datetime(df_new['Timestamp'], dayfirst=True, errors='coerce', format='%d.%m.%Y %H:%M')
```

```python
#Об'єднуємо наші серії, щоб пусті значення в першій серії заповнилися значеннями з другої серії і замінюємо цими даними колоночку
#Timestamp в DataFrame. На виході отримуємо дати в форматі datetime64[ns] з однаковим форматом відображення.
df_new.Timestamp = date_1.fillna(date_2)
df_new.Timestamp
```

```python
#Перевіряємо, чи всі строчки в нас заповнені датами, щоб переконатись, що в нас було тільки 2 різних формату запису дати.
df_new.Timestamp.isna().unique()
```

```python
df_new['Знання англійської мови'] = pd.Categorical(df_new['Знання англійської мови'], 
                      categories=['Elementary','Pre-Intermediate','Intermediate','Upper-Intermediate','Advanced'],
                      ordered=True)

df_new['Ваш тайтл'] = pd.Categorical(df_new['Ваш тайтл'], 
                      categories=['Intern/Trainee', 'Junior', 'Немає тайтлу', 'Middle', 'Senior', 'Tech Lead', 'Team Lead', 'Architect', 'Head', 'Manager'],
                      ordered=True)
```

12. Зробіть зводну таблицю, де стовпці - це 'Ваш тайтл', рядки - 'Знання англійської мови', а значення - медіанна ЗП. Які висновки можна зробити?

```python
pivot_eng_title_salary = df_new.pivot_table(index=['Знання англійської мови'], 
                                            columns=['Ваш тайтл'], 
                                            values=['Зарплата у $$$ за місяць, лише ставка після сплати податків'],
                                            aggfunc='median')

pivot_eng_title_salary.style.format("{:.2f}").background_gradient(axis=0)
```

Для аналізу візьмемо найпоширеніші серед опитаних рівень англійської "Upper-Intermediate" та тайтл "Middle"

```python
#Будуємо сводну таблицю розподілу за статтю по кожній посаді
pivot_position_gender_lang_title = (df_new.pivot_table(index=['Оберіть вашу посаду'], 
                                                       columns=['Ваша стать'], 
                                                       values=['Timestamp'], 
                                                       aggfunc='count'))
pivot_position_gender_lang_title
```

За результатами видно, що, в цілому, чоловіки переважають майже по всіх посадах окрім двох, а саме HR/Recruiter та Marketing, SEO, Copywriter. Найбільша різниця по Software Engineer (Junior, Middle, Senior, Team/Tech Lead, Architect). А в Hardware Engineer не представлена жодна жінка.

```python
#Визначемо функцію, яка порахує середню заробітну плату по кожній з посад для жінок та чоловіків
def mean_salary(position, gender):
    res_w = (df_new[(df_new['Ваша стать'] == gender) 
                   & (df_new['Оберіть вашу посаду'] == position) 
                   & (df_new['Знання англійської мови'] == 'Upper-Intermediate') 
                   & (df_new['Ваш тайтл'] == 'Middle')]['Зарплата у $$$ за місяць, лише ставка після сплати податків'].mean())
    return res_w
```

```python
#Застосувавши функцію mean_salary, формуємо словники, які міститимуть в якості ключів назву посади, а в якості значень середню ЗП для жінок та чоловіків
position_mean_salary_women_dict = {}
position_mean_salary_men_dict = {}
for gen in df_new['Ваша стать'].unique():
    for elem in df_new['Оберіть вашу посаду'].unique():
        res = mean_salary(elem, gen)
        if gen == 'Жінка':
            position_mean_salary_women_dict[elem] = res
        else:
            position_mean_salary_men_dict[elem] = res
```

```python
#На основі отриманих словників створюємо серії, де в якості індексів будуть назви посад, а в якості значень середня ЗП та сортуємо за індексом в алфавітному порядку
position_mean_salary_women_series = pd.Series(position_mean_salary_women_dict).sort_index()
position_mean_salary_men_series = pd.Series(position_mean_salary_men_dict).sort_index()
```

```python
#Об'єднуємо отримані серії в один датафрейм
df_gender_mean_salary_lan_title = pd.DataFrame({'Жінки': position_mean_salary_women_series, 'Чоловіки': position_mean_salary_men_series})
df_gender_mean_salary_lan_title.style.format("{:.2f}")
```

Отримані значення nan в деяких ячейках означають, що по даній посаді відсутні респонденти по одному з критерії початкової фільтрації. Тобто або не представлені в посаді взагалі, або мають рівень знання англійської не 'Upper-Intermediate', або мають тайтл відмінний від 'Middle'. Надалі ми не братимо до уваги посади, які мають хоча б одне значення nan

```python
df_gender_mean_salary_lan_title_notnan = df_gender_mean_salary_lan_title[(df_gender_mean_salary_lan_title['Жінки'].notnull()) 
                                & (df_gender_mean_salary_lan_title['Чоловіки'].notnull())]
df_gender_mean_salary_lan_title_notnan.style.format("{:.2f}")
```

```python
#Виводимо різницю між середніми ЗП жінок та чоловіків на аналогічних посадах
df_gender_mean_salary_lan_title_notnan.diff(axis= 1)['Чоловіки'].astype(dtype=np.int_)
```

За результатами обчислень бачимо, що, за рідким виключенням, чоловіки отримують більшу заробітну платню ніж жінки, перебуваючи на одній і тій же посаді, з однаковим рівнем знання англійської мови та однаковим тайтлом.
