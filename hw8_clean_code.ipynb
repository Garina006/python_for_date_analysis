````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})
````

````python
befor = df.memory_usage(index=False).sum()
````

````python
df[['product_name', 'product_category']] = df[['product_name', 'product_category']].astype({'product_name': 'category', 
                                                                                            'product_category': 'category'}, copy=False)
````

````python
after = df.memory_usage(index=False).sum()
````

````python
print(f"Маємо зниження витрат памʼяті на {round((1-after / befor) * 100, 2)}% !")
````
